```python
# Ans 1: We need custom exception to raise a user define exception from the code. And for this we need the exception 
# class to define the extra methods to the exception which are not in the inbuilts classes.
```

```python
def hierarchy(lst,indentation):
    print("-"*indentation,lst.__name__)
    for i in lst.__subclasses__():
        temp=i.__subclasses__()
        if len(temp)>0:
            hierarchy(i,indentation+3)
```

```python
hierarchy(Exception,1)
```

```python
# Ans 4: LookupError is the base exception class for the errors related to the key or index error when indexing or mapping.
# KeyError is the error occurs when the perticular key is not found in the dictionary.
# E.g. 
try:
    dic={"Key1":1,"Key2":2,"Key3":3}
    print(dic["Key0"])
except KeyError as e:
    print("{} key not found.".format(e))

# IndexError is the error which occurs when the index is out of range in a list or iterable.
# E.G.
try:
    li=[1,2,4,5,63]
    print(li[5])
except IndexError as e:
    print(e)
```
